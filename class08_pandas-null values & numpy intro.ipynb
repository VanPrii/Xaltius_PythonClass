```python
data = pd.read_csv("Bengaluru_House_Data.csv")
data
```

```python
#check a shape of data
data.shape #count of (rows, columns)
#data is variable
```

```python
#check null values from data
data.isnull() #shows in boolean result
#false - data is there
#true - data is not there
```

```python
#check null values according to column wise
data.isnull().sum()
```

```python
#check total null values
data.isnull().sum().sum()
```

```python
#fill null values by using 0
data1 = data.fillna(value=0)
data1
```

```python
data1.isnull().sum().sum() #checking if there are no more null values left after replacing to 0
```

```python
#fill null values by using previous value
data2 = data.fillna(method='pad')
data2
```

```python
data2.isnull().sum().sum() #checking if there are no more null values left after replacing to previous value
```

```python
data2.isnull().sum()
```

```python
#fill null values by using next value
data3 = data.fillna(method='bfill')
data3
```

```python
data3.isnull().sum() #remains one not filled because of the last row has no next value
```

```python
#fill null values with my own value each
data4 = data.fillna({
  'society':'vanessa', #column name, value you want
  'balcony':'lino'
})  #dictionary
data4
```

```python
#convert null values into mean values - statistics
data5 = data.fillna(value=data['balcony'].mean()) #column name, mean() will calculate all the values
data5
```

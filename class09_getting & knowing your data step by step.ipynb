```python
#step 7: which is the most ordered item?
#group by function
#5 display entries as default value

c = chipo.groupby('item_name') #item_name - column name
c = c.sum()
c = c.sort_values(['quantity'], ascending=False) #item mostly ordered, default in ascending
c.head(1) #display only first item, the most ordered
```

```python
#most ordered choice description
d = chipo.groupby('choice_description') #column name
d = d.sum()
d = d.sort_values(['quantity'], ascending=False) #item mostly ordered, default in ascending
d.head()
```

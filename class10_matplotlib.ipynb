```python
#visualizing chipotle data
#step 1: import the necessary libraries

import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter #count data values, how many times data A is repeating, etc.

#set this to graphs open internally
#%matplotlib inline
```

```python
#step 2: import the dataset from this address
#step 3: assign it to a variable called chipo
url = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv'
chipo = pd.read_csv(url, sep='\t')
chipo
```

```python
#step 4: see the first 10 entries
chipo.head(10)
```

```python
#get the series of the names
x = chipo.item_name

#use the Counter class from collections tp create a dictionary with keys and frequency
letter_counts = Counter(x)

#convert dictionary into a dataframe
df = pd.DataFrame.from_dict(letter_counts,orient='index')
df = df[0].sort_values(ascending=True)[45:50] #df[0] to remove 0 index and start on the table from the values

#create the plot
df.plot(kind='bar')

#set the title and labels
plt.xlabel('Items')
plt.ylabel('Number of Times Ordered')
plt.title("Most Ordered Chipotle")

#show the plot
plt.show() #generate the graph
```

```python
#step 6: create a scatterplot with the number of items ordered per order price
#Hint: price should be in the X-axis and Items ordered in the Y-axis

#create the list of prices
chipo.item_price = [float(value[1:-1]) for value in chipo.item_price] #strip the $ sign in for loop because its many

#then groupby the orders and sum
orders = chipo.groupby('order_id').sum()

#create the scatterplot
plt.scatter(x = orders.item_price, y=orders.quantity, s=50, c='green')

#set the title and lables
plt.xlabel('Items')
plt.ylabel('Number of Items Ordered')
plt.title("Most Ordered Chipotle")
plt.ylim(0) #to limit graph to 0, will not go through negative numbers

#show the plot
plt.show() #generate the graph
```

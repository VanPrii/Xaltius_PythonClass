```python
#import library
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#for mac users to import links
import ssl

ssl._create_default_https_context = ssl._create_unverified_context
```

```python
#extract the data from CSV file
dataset = pd.read_csv('Salary_Data.csv')
X = dataset.iloc[:,:-1].values #carries the year of experience column
# [:,:-1] - extract the last column
y = dataset.iloc[:,-1].values #carries the salary column
#iloc - integer location based on index
#training the data - 1/3 part of dataset
```

```python
#splitting the dataset into the training set and test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=1/3, random_state=0)
#X_train, X_test, y_train, y_test - variable only
#1/3 - test data size, you can use any number, but will take a little sample only, not standard
#random_state = 0 - we dont want random numbers
```

```python
#training the simple linear regression model on the TRAINING set
from sklearn.linear_model import LinearRegression
regression = LinearRegression() #y = bo + b1X1 automatic calculation
regression.fit(X_train,y_train) #pass the train set
```

```python
#predicting the test result
y_pred = regression.predict(X_test) #prediction will be done on experience
#if sample is good quality or not
```

```python
#visualizing the TRAINING set results
#we need to compare our sample to actual product
plt.scatter(X_train,y_train,color='red')
plt.plot(X_train,regression.predict(X_train), color='blue')
plt.title("Salary vs Experience (Training Set)")
plt.xlabel("Year of Experience")
plt.ylabel("Salary")
plt.show()
```

```python
#visualizing the TEST set results
plt.scatter(X_test,y_test,color='red')
plt.plot(X_train,regression.predict(X_train), color='blue') #line should not be change
plt.title("Salary vs Experience (Test Set)")
plt.xlabel("Year of Experience")
plt.ylabel("Salary")
plt.show()
```
